# TD1 : Préparation du Corpus

L’objectif de ce TD est de préparer ce corpus d’articles en vue de son indexation. Pour réaliser
l’indexation, vous allez construire un fichier, qui rassemble tous les articles dans une structure XML qui
sera facilement indexable. Ce fichier XML devra être structuré de façon à pouvoir retrouver (quand ils
existent) :
- le numéro de l’article,
- le numéro du bulletin,
- la date,
- la rubrique,
- le titre de l’article,
- l’auteur de l’article,
- le texte de l’article,
- la ou les images avec leur(s) URL(s) et leur(s) légende(s) respective(s),
- les informations de contact.
Après avoir téléchargé et « dézipé » ce corpus, vous obtiendrez le répertoire « BULLETINS ».
On a choisi de structurer ce fichier au format XML correspondant à l’arborescence donnée en annexe.

```python
from bs4 import BeautifulSoup
from pathlib import Path
from lxml import etree
```

```python
path = Path("../data/BULLETINS/67385.htm")
with open(path) as f:
    content = f.read()
    soup = BeautifulSoup(content, 'html.parser')
    
# Convert to etree for XPath
dom = etree.HTML(str(soup))

# Extract heading using XPath
title = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[3]/td[1]/p[1]/span[2]/text()')[0]
num_buletin = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[6]/td[3]/p/a/span/text()')[0]
date = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[1]/td[3]/p/span[3]/text()')[0]
```

```python
class Bulletins:
    def __init__(self, path):
        self.path = path
        self.soup = None
        self.dom = None
        self.title = None
        self.num_buletin = None
        self.date = None
        
    def load(self):
        with open(self.path) as f:
            self.content = f.read()
            self.soup = BeautifulSoup(self.content, 'html.parser')
            self.dom = etree.HTML(str(self.soup))
            
    def extract_data(self):
        self.title = self.dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[3]/td[1]/p[1]/span[2]/text()')[0]
        self.num_buletin = self.dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[6]/td[3]/p/a/span/text()')[0]
        self.date = self.dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[1]/td[3]/p/span[3]/text()')[0]
        
    def makeXML(self):
        root = etree.Element("Bulletin")
        title_elem = etree.SubElement(root, "Title")
        title_elem.text = self.title
        num_elem = etree.SubElement(root, "NumBulletin")
        num_elem.text = self.num_buletin
        date_elem = etree.SubElement(root, "Date")
        date_elem.text = self.date
        return etree.tostring(root, pretty_print=True).decode()
```
