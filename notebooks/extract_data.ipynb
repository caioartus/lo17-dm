```python
from bs4 import BeautifulSoup
from pathlib import Path
from lxml import etree
import re
```

```python
path = Path("../data/BULLETINS/67385.htm")
with open(path) as f:
    content = f.read()
    soup = BeautifulSoup(content, 'html.parser')
```

```python
# Convert to etree for XPath
dom = etree.HTML(str(soup))

# Extract heading using XPath
title = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[3]/td[1]/p[1]/span[2]/text()')[0].strip()
auteur_info = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[8]/td[2]/p/span/text()')[0].strip()
auteur_email = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[8]/td[2]/p/span/a/text()')[0]
num_article = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[6]/td[3]/p/a/span/text()')[0]
num_buletin = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[1]/td[3]/p/span[1]/text()')[0].strip().replace("BE France", "").strip()
date = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[1]/td[3]/p/span[3]/text()')[0].strip()
rubrique = dom.xpath('//*[@id="LayoutTable"]/table/tr[7]/td/table/tr[3]/td[1]/p[1]/span[1]/text()')[0].strip()
```
